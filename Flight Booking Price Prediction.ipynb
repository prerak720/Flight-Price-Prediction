```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
   
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data.set_index('Feature')['VIF']

```

### Liner Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['price','flight'])
y = df[['price']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=33)
```

```python
li = LinearRegression()
```

```python
model = li.fit(x_train,y_train)
```

```python
predication = model.predict(x_test)
```

```python
r2_score(predication,y_test)
```

```python
mean_absolute_error(predication,y_test)
```

```python
mean_squared_error(predication,y_test)
```

```python
sns.distplot(y_test,label='Actuar')
sns.distplot(predication,label = 'Prediction')
plt.legend()
```

here Lineregression get 89 % accurancy

### LogicalRegression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
di = DecisionTreeRegressor()
```

```python
model_1 = di.fit(x_train,y_train)
```

```python
prediction_1 = model_1.predict(x_test)
```

```python
print(f'mean_absolute_error {mean_absolute_error(prediction_1,y_test)}')
print(f'mean_squared_error {mean_squared_error(prediction_1,y_test)}')
print(f'r2_score {r2_score(prediction_1,y_test)}')
```

```python
sns.distplot(y_test,label='Actuar')
sns.distplot(prediction_1,label = 'Prediction')
plt.legend()
```

The acuurancy is high compared to linear regression.

96% accuracy achieved with the Decision TreeML algorithm.
